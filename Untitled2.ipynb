```python
def jumlah_dataset(directory):
  jumlah = 0
  for path, subdirs, files in os.walk(directory):
    for name in files:
      jumlah += 1
  return jumlah
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen= ImageDataGenerator(
                    rescale=1./255,
                    rotation_range=20,
                    width_shift_range=0.1,
                    height_shift_range=0.1,
                    horizontal_flip=True,
                    vertical_flip=True,
                    brightness_range=[0.8,1.3],
                    shear_range = 0.2,
                    fill_mode = 'nearest'
                    )

test_datagen = ImageDataGenerator(
                    rescale=1./255)
```

```python
traingen= train_datagen.flow_from_directory(
    train_dir,
    batch_size=64,
    target_size=(224,224),
    class_mode='categorical'
)
valgen=test_datagen.flow_from_directory(
    test_dir,
    batch_size=64,
    class_mode='categorical'
)
```

```python
from tensorflow.keras.applications import MobileNetV2
base_model = MobileNetV2(weights='imagenet', include_top=False)
```

```python
for layer in base_model.layers:
    layer.trainable = False
```

```python
from keras.layers import Input, BatchNormalization
from keras.regularizers import l2
model = Sequential([
    Input(shape=(224,224,3)),
    base_model,
    Conv2D(16, (3,3), padding='same', activation='relu'),
    MaxPooling2D(2,2, padding='same'),
    BatchNormalization(),
    Flatten(),
    Dense(100, activation='relu', kernel_regularizer=l2),
    Dropout(0.5),
    Dense(3, activation='softmax')
])
print(model.summary())
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
earlystop = EarlyStopping(monitor='val_accuracy', patience=3, verbose=1, restore_best_weights=True)
class akurasistop(Callback):
    def on_epoch_end(self, epoch, logs={}):
        if(logs.get('accuracy') >0.95 and logs.get('val_accuracy')>0.95):
            print("\nAkurasi telah mencapai >95%!")
            self.model.stop_training = True
```

```python
history = model.fit(
    traingen,
    epochs=30,
    validation_data=valgen,
    callbacks=[akurasistop(), earlystop]
    )
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Akurasi Model')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Loss Model')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
```

```python
base_model.trainable = True
fine_tune_at = 15  # Fine-tuning dari lapisan ke-15

for layer in base_model.layers[:fine_tune_at]:
    layer.trainable = False

model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

fine_tune_history = model.fit(
    traingen,
    validation_data=valgen,
    epochs=20,
    callbacks=[akurasistop()]
)
```

```python
plt.plot(fine_tune_history.history['accuracy'])
plt.plot(fine_tune_history.history['val_accuracy'])
plt.title('Akurasi Model')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
```

```python
import numpy as np
from keras.preprocessing.image import load_img, img_to_array
from keras.models import load_model

# 2. Memuat dan memproses gambar
def preprocess_image(image_path, target_size):
    # Memuat gambar
    img = load_img(image_path, target_size=target_size)
    # Mengubah gambar menjadi array NumPy
    img_array = img_to_array(img)
    # Menambahkan dimensi batch (model memerlukan dimensi batch, walaupun hanya satu gambar)
    img_array = np.expand_dims(img_array, axis=0)
    # Normalisasi gambar jika diperlukan (misalnya, jika model dilatih dengan gambar dinormalisasi)
    img_array = img_array / 255.0
    return img_array

# Ganti dengan path gambar yang ingin diprediksi dan ukuran target sesuai input model
image_path = 'image.jpg'
target_size = (224, 224)  # Ganti dengan ukuran input model Anda
img_array = preprocess_image(image_path, target_size)

# 3. Melakukan prediksi
predictions = model.predict(img_array)

# 4. Menginterpretasikan hasil
# Mengambil indeks kelas dengan probabilitas tertinggi
class_index = np.argmax(predictions[0])

# Mendapatkan nama kelas dari indeks
class_labels = ['class_1', 'class_2', 'class_3']  # Ganti dengan label kelas yang sesuai
predicted_class = class_labels[class_index]

print(f'Predicted class: {predicted_class}')
```
